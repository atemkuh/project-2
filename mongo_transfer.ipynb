```python
import csv
import json
import pymongo
import pandas as pd
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.test_db
```

<h3> Loading 12_year accident collection

```python
csvfile = open('csv/12_year_accident.csv', 'r')
```

```python
reader = csv.DictReader(csvfile)
```

```python
header = ['Date', 'Type', 'Registration', 'Operator', 'Fatalities', 'Location', 'Latitude', 'Longitude']
```

```python
for each in reader:
    row = {}
    for field in header:
        row[field] = each[field]
#twelve_year is a collection within the database test_db
    db.twelve_year.insert(row)
```

```python
with client:

    db = client.test_db

    ops = db.twelve_year.find()

    for op in ops:
        print('{0} {1}'.format(op['Date'], 
            op['Fatalities']))
```
